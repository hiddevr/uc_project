```python
import pandas as pd

df = pd.read_csv('escooter_data.csv')
df = df.dropna(subset=['Start Community Area Number', 'End Community Area Number'])
df
```

```python
import numpy as np

start_areas = df['Start Community Area Number'].unique()
end_areas = df['End Community Area Number'].unique()
all_areas = np.unique(np.concatenate((start_areas, end_areas)))

area_to_index = {area: idx for idx, area in enumerate(all_areas)}

transition_matrix = np.zeros((len(all_areas), len(all_areas)))

for index, row in df.iterrows():
    start_idx = area_to_index[row['Start Community Area Number']]
    end_idx = area_to_index[row['End Community Area Number']]
    transition_matrix[start_idx, end_idx] += 1

for i in range(len(all_areas)):
    total_trips_from_area = transition_matrix[i, :].sum()
    if total_trips_from_area > 0:
        transition_matrix[i, :] /= total_trips_from_area

transition_df = pd.DataFrame(transition_matrix, 
                             index=all_areas, 
                             columns=all_areas)
transition_df
```

```python
transition_df.to_csv('transition_matrix.csv')
```
